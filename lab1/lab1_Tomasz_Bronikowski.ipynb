```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl

from sklearn.datasets import load_iris
irisRaw = load_iris()

iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

Zadanie 8

```python
min = iris["petal length (cm)"].min()
min_id = iris["petal length (cm)"].idxmin()
max = iris["petal length (cm)"].max()
max_id = iris["petal length (cm)"].idxmax()
print('Min = ',min, 'id = ',min_id)
print('Max = ',max, 'id = ',max_id)
```

Zadanie 11

```python
iris.loc[iris['target'] == 0].hist(column=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'], figsize=(10,6))
pl.suptitle("Target 0")
plt.show()

iris.loc[iris['target'] == 1].hist(column=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'], figsize=(10,6))
pl.suptitle("Target 1")
plt.show()

iris.loc[iris['target'] == 2].hist(column=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'], figsize=(10,6))
pl.suptitle("Target 2")
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine

data = load_wine()
data_df = pd.DataFrame(data.data, columns = data.feature_names)
```

```python
data_df.head()
```

```python
data_df.describe()
```

```python
from sklearn.model_selection import train_test_split, cross_val_score

X_train, X_test, y_train, y_test = train_test_split(data_df, data.target, test_size=.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

decisionTree = DecisionTreeClassifier(min_samples_leaf=3, random_state=1)
decisionTree = decisionTree.fit(data_df, data.target)
decisionTreeValues = cross_val_score(decisionTree, data_df, data.target)

acc = metrics.accuracy_score(y_test, decisionTree.predict(X_test))
expected = [0.94444444, 0.80555556, 0.80555556, 0.91428571, 0.97142857]

print(acc)
print(decisionTreeValues)
print(expected)
```

```python
from sklearn.ensemble import BaggingClassifier

bagging = BaggingClassifier(estimator = DecisionTreeClassifier(min_samples_leaf=3, random_state=1), random_state=1, n_estimators=70)
bagging = bagging.fit(data_df, data.target)
baggingValues = cross_val_score(bagging, data_df, data.target)

acc = metrics.accuracy_score(y_test, bagging.predict(X_test))
expected = [0.91666667, 0.91666667, 0.94444444, 0.97142857, 1.]

print(acc)
print(baggingValues)
print(expected)
```

```python
from sklearn.ensemble import AdaBoostClassifier

adaBoost = AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1, min_samples_leaf=3, random_state=1), random_state=1, n_estimators=70, algorithm='SAMME')
adaBoost = adaBoost.fit(data_df, data.target)
adaBoostValues = cross_val_score(adaBoost, data_df, data.target)

acc = metrics.accuracy_score(y_test, adaBoost.predict(X_test))
expected =  [0.83333333, 0.91666667, 0.91666667, 1.,1.]

print(acc)
print(adaBoostValues)
print(expected)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gradientBoosting = GradientBoostingClassifier(n_estimators=70, max_depth=1, learning_rate=1, random_state=1, subsample=0.5, min_samples_leaf=3)
gradientBoosting = gradientBoosting.fit(data_df, data.target)
gradientBoostingValues = cross_val_score(gradientBoosting, data_df, data.target)



acc = metrics.accuracy_score(y_test, gradientBoosting.predict(X_test))
expected= [0.91666667, 0.97222222, 0.97222222, 0.42857143, 0.97142857]

print(acc)
print(gradientBoostingValues)
print(expected)
```

```python
cumsum = np.cumsum(gradientBoosting.oob_improvement_)
x_val = np.linspace(0,len(cumsum), 70)
plt.plot(x_val, cumsum)
plt.show()
```
